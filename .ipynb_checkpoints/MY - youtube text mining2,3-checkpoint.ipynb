```python
comment_data = pd.read_csv('text_mining.csv')
```

```python
#ì´ëª¨í‹°ì½˜ ì œê±°
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)

#ë¶„ì„ì— ì–´ê¸‹ë‚˜ëŠ” ë¶ˆìš©ì–´êµ¬ ì œì™¸ (íŠ¹ìˆ˜ë¬¸ì, ì˜ì„±ì–´)
han = re.compile(r'[ã„±-ã…ã…-ã…£!?~,".\n\r#\ufeff\u200d]')
```

```python
comment_list = []
for i in range(len(comment_data)):
    comment_list.append(comment_data['comment'].iloc[i])
```

```python
comment_result = []

for i in comment_list:
    tokens = re.sub(emoji_pattern,"",i)
    tokens = re.sub(han,"",tokens)
    comment_result.append(tokens)

comment_result = pd.DataFrame(comment_result, columns=["comment"])
```

```python
from konlpy.tag import Twitter

def get_noun(comment_txt):
    twitter = Twitter()
    noun = []
    
    if len(comment_txt)>0:
        tw = twitter.pos(comment_txt)
        for i,j in tw:
            if j == 'Noun':
                noun.append(i) #ëª…ì‚¬ì¸ ë‹¨ì–´ë§Œ ì¶”ì¶œ
    return noun
 
comment_result['token'] = comment_result['comment'].apply(lambda x: get_noun(x))
```

```python
vec = []
for i in range(len(comment_result)):
    vec2 = []
    tm_ls = comment_result['token'].iloc[i]
    if len(tm_ls) == 0: #ë¹„ì–´ìˆëŠ” ë¦¬ìŠ¤íŠ¸ ì‚­ì œ
        pass
    else:
        for j in range(len(tm_ls)): #ë¹„ì–´ìˆì§€ ì•Šì€ ë¦¬ìŠ¤íŠ¸ ì¤‘ì—ì„œ ë‹¨ì–´ê°€ í•œ ìŒì ˆì¸ ê²ƒì€ pass
            if len(tm_ls[j]) > 1:
                vec2.append(tm_ls[j])
            else:
                pass
    if len(vec2) == 0:
        pass
    else:
        vec.append(vec2)
        
```

```python
model = Word2Vec(vec, min_count=3,window=3,iter=20, size=100, sg=1) #skip-gram
```

```python
from sklearn.cluster import KMeans
```

```python
model.wv.syn0.shape #w2v size  -> 45ê°œì˜ ë‹¨ì–´ê°€ 100ì°¨ì›ì„ ê°€ì§€ê³  ìˆìŒ

#45ê°œì˜ ë‹¨ì–´ê°€ 100ì°¨ì›ìœ¼ë¡œ ë²¡í„°í™”ëœ ìˆ˜ì¹˜ë¥¼ ë„£ì–´ì¤Œ
word_vectors = model.wv.syn0 # model.wv.vectorsì™€ ê°™ì€ ê°’ì„ ê°€ì§
```

```python
word_vectors
```

```python
num_clusters = 6 #clusterì˜ ê°¯ìˆ˜ ì‚°ì •
kmeans_clustering = KMeans( n_clusters = num_clusters )#default
```

```python
idx = kmeans_clustering.fit_predict( word_vectors ) #ê° ë‹¨ì–´ê°€ ì–´ë–¤ êµ°ì§‘ì— ì†í•˜ëŠ”ì§€ í™•ì¸
```

```python
idx
```

```python
# ê° ì–´íœ˜ ë‹¨ì–´ë¥¼ í´ëŸ¬ìŠ¤í„° ë²ˆí˜¸ì— ë§¤í•‘ë˜ê²Œ word/Index ì‚¬ì „ì„ ë§Œë“ ë‹¤.
idx = list(idx)
names = model.wv.index2word #í•™ìŠµëœ ë‹¨ì–´ë“¤ unique
word_centroid_map = {names[i]: idx[i] for i in range(len(names))}#ê° ë‹¨ì–´ ë³„ë¡œ í´ëŸ¬ìŠ¤í„° ë²ˆí˜¸ì— ë§¤ì¹­
```

```python
for cluster in range(0,6):
    # í´ëŸ¬ìŠ¤í„° ë²ˆí˜¸ë¥¼ ì¶œë ¥
    print("\nCluster {}".format(cluster))
    
    # í´ëŸ¬ìŠ¤í„°ë²ˆí˜¸ì™€ í´ëŸ¬ìŠ¤í„°ì— ìˆëŠ” ë‹¨ì–´ë¥¼ ì°ëŠ”ë‹¤.
    words = []
    for i in range(0,len(list(word_centroid_map.values()))): #list(dict)í‚¤ ê°’ë§Œ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
        #list(dict.values)ëŠ” valueê°’ì´ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜
        if( list(word_centroid_map.values())[i] == cluster ):
            words.append(list(word_centroid_map.keys())[i])
    print(words)
```
