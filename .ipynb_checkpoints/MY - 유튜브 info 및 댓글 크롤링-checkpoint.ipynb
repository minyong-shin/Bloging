ì´ëŸ¬í•œ ë°©ë²•ìœ¼ë¡œ êµ°ì§‘í™”ë¥¼ í•˜ê²Œ ëœë‹¤ë©´ í´ëŸ¬ìŠ¤í„°ì˜ ê°¯ìˆ˜ë¥¼ ì§€ì •í•  í•„ìš”ê°€ ì—†ìœ¼ë©° ì›í˜•ì´ ì•„ë‹Œ í´ëŸ¬ìŠ¤í„° í˜•íƒœë¡œë„ ì¶”ì¶œì´ ê°€ëŠ¥í•˜ë‹¤ëŠ” ì ì´ë‹¤.

dbscanì€ ê°ê° ê°œë³„ í¬ì¸íŠ¸ ê°„ì˜ ê±°ë¦¬ë¥¼ ì¸¡ì •í•˜ë©´ì„œ **non-core-point**ì— ì˜í•´ì„œ í´ëŸ¬ìŠ¤í„°ì˜ ê²½ê³„ê°€ í˜•ì„±ë˜ê¸° ë•Œë¬¸ì— ì›í˜•ì´ ì•„ë‹Œ ëª¨ì–‘(ê¸°í•˜í•™ì  ëª¨ì–‘)ì˜ êµ°ì§‘í™”ê°€ ê°€ëŠ¥í•œ ê²ƒì„

### ë³´ëŠ” ê²ƒê³¼ ê°™ì´ ì í•©í•œ epsilonê³¼ min_samplesë¥¼ ì°¾ëŠ” ê²ƒì´ dbscanì˜ í•µì‹¬ì´ë‹¤  
### dbscanì´ ì–´ë– í•œ ë°©ì‹ìœ¼ë¡œ í˜•ì„±ë˜ê³  ì–´ë–»ê²Œ ì§„í–‰ ë˜ëŠ”ì§€ëŠ” ì´í•´í•˜ê³  ìˆì§€ë§Œ ìµœì ì˜ epsilonê³¼ min_sampleë¥¼ ì°¾ëŠ” ê²ƒì€ ì—°ìŠµì´ í•„ìš”í•¨ ë°‘ê³¼ ìœ„ì˜ ì½”ë“œëŠ” ì£¼ì„±ë¶„ë¶„ì„ê³¼ nmfì„±ë¶„ìœ¼ë¡œ ë‚˜ëˆ ì„œ ë”°ë¡œë”°ë¡œ ë¶„ì„ì„ ì§„í–‰í•˜ê³  ìµœì ì˜ ê°’ì„ ì°¾ì€ ê²ƒì´ë‹¤ í•´ë‹¹ ì½”ë“œë“¤ì€ ê¼­ ì´í•´í•  ê²ƒ!

```python
for min_sample in [3, 5, 7, 9]:

    for eps in [0.2, 0.4, 0.9, 1, 3, 5, 7, 9, 11, 13, 15, 17]:

        dbscan = DBSCAN(n_jobs=-1, min_samples=min_sample, eps=eps)# -1ì„ í†µí•´ì„œ ë…¸ì´ì¦ˆ ì œê±°

        labels_pca = dbscan.fit_predict(x_pca) # pcaë¥¼ ì´ìš©í•œ ì„±ë¶„ì„ ì´ìš©í•œ dbscanì„ ì˜ˆì¸¡

        labels_nmf = dbscan.fit_predict(x_nmf) # nmfë¥¼ ì´ìš©í•œ ì„±ë¶„ì„ ì´ìš©í•œ dbscanìœ¼ë¡œ ì˜ˆì¸¡

        print('ï¸=== min_sample:{}, eps:{} ==='.format(min_sample, eps))

        print('')

        print('pca ==> cluster ìˆ˜: {}'.format(len(np.unique(labels_pca))))

        print('pca ==> cluster í¬ê¸°: {}\n'.format(np.bincount(labels_pca+1)))



        print('nmf ==> cluster ìˆ˜: {}'.format(len(np.unique(labels_nmf))))

        print('nmf ==> cluster í¬ê¸°: {}'.format(np.bincount(labels_nmf+1)))

        print('---------------------------------------')
```
